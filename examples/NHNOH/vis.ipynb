```python
from pathlib import Path

import yaml

import autochem
from mess2nasa.cli import Input

input_file = "input.yaml"
```

```python
# 1. Read input
input_data = yaml.safe_load(Path(input_file).read_text())
input = Input.model_validate(input_data)
formula = input.formula
Hf = input.Hf
Tf = input.Tf
units = {"energy": input.energy_unit}
mess_inp_name = Path(input.mess_input)
nasa_out_name = Path(input.nasa_output)

# 2. Read MESS-PF output
pf_str = mess_inp_name.with_suffix(".dat").read_text()
therm_pf = autochem.therm.from_messpf_output_string(
    pf_str, formula=formula, Hf=Hf, Tf=Tf, units=units
)

# 3. Read fit
fit_str = nasa_out_name.read_text()
therm_fit = autochem.therm.from_chemkin_string(fit_str)

autochem.therm.display(therm_pf, others=[therm_fit], label="raw", others_labels=["fit"])
```
